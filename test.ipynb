```python
import pandas as pd
dir = 'Record Collections/ConsineHashMap.csv'
df = pd.read_csv(dir, encoding='utf-8', header=None)

hashmap = {}
for i in range(df.shape[1]):
    # 第一行为key
    # 下面的行为value
    key = df.iloc[0, i]
    value = df.iloc[1:, i].dropna().tolist()
    hashmap[key] = value

print(hashmap.keys())
```
